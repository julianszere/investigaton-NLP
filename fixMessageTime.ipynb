```python
import os
import pandas as pd
from src.datasets.LongMemEvalDataset import LongMemEvalDataset

dataset = LongMemEvalDataset("short", "longmemeval")

for instance in dataset:
    cache_path = f"data/rag/{instance.question_id}.parquet"
    if not os.path.exists(cache_path):
        continue

    df = pd.read_parquet(cache_path)
    if "messages_time" in df.columns:
        continue  # already migrated

    messages_time = []
    for session in instance.sessions:
        session_time = session.date
        for _ in session.messages:
            messages_time.append(session_time)

    if len(messages_time) != len(df):
        print("Length mismatch for", instance.question_id, len(messages_time), len(df))
        continue

    df["messages_time"] = messages_time
    df.to_parquet(cache_path)
    print("Migrated", cache_path)
```
