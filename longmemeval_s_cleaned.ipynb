```python
import json
import pandas as pd
```

```python
with open('longmemeval-cleaned/Long Memeval Cleaned.json', 'r', encoding='utf-8') as f:
    longmemeval_data = json.load(f)
```

```python
len(longmemeval_data)
```

Explicación breve de cada columna del dataset:

- **question_id**
- **question_type**: Tipo de pregunta, puede ser XXXX
- **question**
- **question_date**: fecha y hora en que fue hecha la pregunta.
- **answer**
- **answer_session_ids**: lista de IDs de sesiones relevantes donde aparece la respuesta correcta.
- **haystack_dates**: fechas y horas asociadas a los mensajes de la historia/haystack.
- **haystack_session_ids**: identificadores de las sesiones de la historia/haystack.
- **haystack_sessions**: estructura anidada con el contenido de los diálogos previos a la pregunta (historial conversacional).
